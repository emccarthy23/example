Defining Sorting Functions

```python
def quicksort(x):
    """
    Describe how you are sorting `x`
    Choose a pivot. Arrange all the elements so the larger elements are in front of the pivot
    and the smaller elements are behind the pivot. Then recursively apply the funciton on the sublists.
    """
    quicksort_recursive(x,0,len(x)-1)
    return x

def quicksort_recursive(x,low,high):
    if high > low:

        pivot = partition(x,low,high)
        quicksort_recursive(x,low,pivot-1)
        quicksort_recursive(x,pivot+1,high)
        
def partition(x,low,high):
    
    middle = x[low]
    lower_index = low
    higher_index = high
    
    while higher_index > lower_index:
        
        while lower_index <= higher_index and x[lower_index] <= middle:
            lower_index = lower_index + 1
            
        while x[higher_index] >= middle and higher_index >= lower_index:
            higher_index = higher_index - 1
            
        if higher_index > lower_index:
            x[higher_index],x[lower_index] = x[lower_index],x[higher_index]       
    x[low],x[higher_index] = x[higher_index],middle
    return higher_index

    
```

```python
def bubblesort(x):
    """
    Describe how you are sorting `x`:
    Start at the beginning of the list and for each pair of elements
    i and i+1 flip the order if i+1 is smaller than i.
    Iterate over the list pairwise until the list is sorted.
    """
    if (len(x) > 1):
        j=1
        while j!=0:
            j=0
            for i in range(len(x)-1):    
                if x[i] > x[i+1]:
                    x[i],x[i+1] = x[i+1],x[i]
                    j = j+1
        return x
    else:
        return x
```

Defining functions with conditional and assignment counters

```python
def bubblesort(x):
    """
    Describe how you are sorting `x`:
    Start at the beginning of the list and for each pair of elements
    i and i+1 flip the order if i+1 is smaller than i.
    Iterate over the list pairwise until the list is sorted.
    """
    conditional = 0
    assignment = 0
    conditional = conditional + 1
    if (len(x) > 1):
        j=1
        assignment = assignment + 1
        
        
        while j!=0:
            conditional = conditional + 1
            j=0
            assignment = assignment + 1
            
            for i in range(len(x)-1):
                conditional = conditional + 1
                if x[i] > x[i+1]:
                    x[i],x[i+1] = x[i+1],x[i]
                    assignment = assignment + 1
                    j = j+1
                    assignment = assignment + 1
        conditional = conditional + 1
        return x, conditional, assignment
    else:
        return x, conditional, assignment
```

```python
def quicksort(x):
    """
    Describe how you are sorting `x`
    Choose a pivot. Arrange all the elements so the larger elements are in front of the pivot
    and the smaller elements are behind the pivot. Then recursively apply the funciton on the sublists.
    """
    conditional, assignment = quicksort_recursive(x,0,len(x)-1)
    return x, conditional, assignment
    


def quicksort_recursive(x,low,high):
    assignment = 0
    conditional=0
    conditional=conditional+1
    if high > low:
        
        pivot, cond_add, assign_add, = partition(x,low,high)
        conditional = conditional + cond_add
        assignment = assignment + assign_add
        cond_add_1, assign_add_1 = quicksort_recursive(x,low,pivot-1)
        conditional = conditional + cond_add_1
        assignment = assignment + assign_add_1
        cond_add_2, assign_add_2 = quicksort_recursive(x,pivot+1,high)
        conditional = conditional + cond_add_2
        assignment = assignment + assign_add_2
        return conditional, assignment
    return conditional, assignment

        
def partition(x,low,high):
    assignment = 0
    conditional = 0
    middle = x[low]
    assignment = assignment + 1
    lower_index = low
    assignment = assignment + 1
    higher_index = high
    assignment = assignment + 1
    
    while higher_index > lower_index:
        conditional = conditional + 1
        while lower_index <= higher_index and x[lower_index] <= middle:
            conditional = conditional + 1
            lower_index = lower_index + 1
            assignment = assignment + 1
        conditional = conditional + 1    
        while x[higher_index] >= middle and higher_index >= lower_index:
            conditional = conditional + 1
            higher_index = higher_index - 1
            assignment = assignment + 1
            
        conditional = conditional + 1
        
        conditional = conditional + 1
        if higher_index > lower_index:
            x[higher_index],x[lower_index] = x[lower_index],x[higher_index]
            assignment = assignment + 1
    conditional = conditional + 1
    x[low],x[higher_index] = x[higher_index],middle
    assignment = assignment + 1
    return higher_index, conditional, assignment
```
